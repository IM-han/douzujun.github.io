### Seaborn 整体布局风格设置
- 单变量的关系
- 多变量的关系

```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<img src="f1.png" alt="FAO" width="990">

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# 没有使用 seaborn之前
sinplot()
```

```python
# 用seaborn默认的组合, 设置了参数之后
sns.set()
sinplot()
```

### 5种主题风格 ###
* darkgrid
* whitegrid
* dark
* white
* ticks

#### sns.set_style("whitegrid") 

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
# 
sns.despine()
```

### 风格细节设置

```python
#f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
# 隐藏指定的轴线
# sns.despine(left=True)
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.set()
```

```python
sns.set_context("paper")
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("talk")
plt.figure(figsize=(8, 6))
sinplot()
```
